```python
import os
import scipy.io
import matplotlib.pyplot as plt

# Paths
mat_dir = "/home/stevenyang/Desktop/cmput469/YOLO/NWPU_60/mats"

def plot_people_distribution():
    num_people_list = []

    mat_files = sorted(os.listdir(mat_dir))
    for mat_file in mat_files:
        if not mat_file.endswith(".mat"):
            continue
        
        mat_path = os.path.join(mat_dir, mat_file)
        mat_data = scipy.io.loadmat(mat_path)
        ann_boxes = mat_data['annBoxes']
        
        num_people_list.append(len(ann_boxes))

    plt.figure(figsize=(10, 6))
    plt.hist(num_people_list, bins=50, color="blue", alpha=0.7, edgecolor="black")
    plt.xlabel("Number of People in Image")
    plt.ylabel("Frequency")
    plt.title("Distribution of Number of People per Image")
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()

# Run the function
plot_people_distribution()
```
